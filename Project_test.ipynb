```python
data_dop = day_data.dropna(axis=0)
data_dop
```

```python
data_dop = data_dop.loc[:,['거래량','매출액', 'PER','부채비율','전일비','종가']]
data_dop
```

```python
# MinMaxScaler : 최대/최소값이 각각 1, 0이 되도록 스케일링
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()#최소 최대 지정(0과1이 기본)
scaler.fit(data_dop.iloc[:,:-1])
X_data = scaler.transform(data_dop.iloc[:,:-1])
X_data
```

```python
# 데이터 분할하기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_data, data_dop.iloc[:,-1], shuffle=False)
X_train
```

```python
# 모델 생성 및 학습
from sklearn.linear_model import LinearRegression
model = LinearRegression() #coef_가 w, intercept_가 b
model.fit(X_train, y_train)
```

```python
# 모델 예측 및 평가
score = model.score(X_test, y_test)
print("정확도 :", score)
cofficient = model.coef_
intercept = model.intercept_
print("계수 :", cofficient)
print("절편 :", intercept)
```

```python
# evaluate an ARIMA model using a walk-forward validation
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot
from statsmodels.tsa.arima.model import ARIMA
from math import sqrt
from sklearn.metrics import mean_squared_error
X = show_data.values
size = int(len(X) * 0.8)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
# walk-forward validation
for t in range(len(test)):
    model = ARIMA(history,order=(3,0,2))
    model_fit = model.fit()
    output = model_fit.forecast()
    #print(model_fit.forecast(steps=3))
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)   #test의 데이터를 집어넣어 한 단계 바로 뒤만을 예측함
    #print('predicted=%f, expected=%f' % (yhat, obs))
# evaluate forecasts
rmse = sqrt(mean_squared_error(test, predictions))
model = ARIMA(history,order=(5,1,0))
output = model.fit().forecast()
print('다음날 시가 예측 : %f'%(output))
print('Test RMSE: %.3f' % rmse)
# # plot forecasts against actual outcomes
pyplot.plot(test)
pyplot.plot(predictions, color='red')
pyplot.show()
```
